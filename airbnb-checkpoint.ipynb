```python
url = 'https://raw.githubusercontent.com/rosemutai/Airbnb-Dataset-in-Kenya/refs/heads/main/airbnb.json'
```

```python
import requests
```

```python
data_content = requests.get(url)
```

```python
json_data = data_content.json()
```

```python
json_data[:5]
```

```python
import pandas
```

```python
data_df = pandas.DataFrame(json_data)
```
